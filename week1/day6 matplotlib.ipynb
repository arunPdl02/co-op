```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Line Plot - Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({'x': x, 'y': y})
sns.lineplot(data=df, x='x', y='y')
plt.title("Seaborn Line Plot")
plt.show()
```

```python
## bar plot
```

```python
categories = ['A', 'B', 'C']
values = [4, 7, 1]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.ylabel("Values")
plt.show()
```

```python
df = pd.DataFrame({'Category': categories, 'Value': values})
sns.barplot(data=df, x='Category', y='Value')
plt.title("Seaborn Bar Chart")
plt.show()
```

```python
df = pd.DataFrame({
    'Year': [2019, 2020, 2021, 2022, 2023, 2024],
    'Sales' : np.random.randint(400000,600000, 6),
    'Profit' : np.random.randint(100000, 250000, 6)
})
print(df)
```

```python
sns.lineplot(data=df, x='Year', y='Sales', marker='o')
plt.title("Line Plot of Sales Over Years")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.show()
```

```python
# Melt for grouped bar plot
df_melted = df.melt(id_vars='Year', value_vars=['Sales', 'Profit'], 
                    var_name='Metric', value_name='Amount')

sns.barplot(data=df_melted, x='Year', y='Amount', hue='Metric')
plt.title("Sales vs Profit per Year")
plt.ylabel("Amount ($)")
plt.show()

## id_vars is what variables to keep, value_vars is what variables to melt into one, var_name is the name of the new col with the old col vals
## and value_name is new col with old values
```

```python
plt.scatter(df['Year'], df['Sales'], color='blue', label='Sales')
plt.scatter(df['Year'], df['Profit'], color='red', label='Profit')
plt.title("Scatter Plot: Sales vs Profit Over Years")
plt.xlabel("Year")
plt.ylabel("Amount ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_melted = df.melt(id_vars='Year', value_vars=['Sales', 'Profit'], 
                    var_name='Metric', value_name='Amount')

sns.scatterplot(data=df_melted, x='Year', y='Amount', hue='Metric')
plt.grid(True)
plt.show()
```

```python
### Facetting
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
# Facet by 'sex'
g = sns.FacetGrid(tips, col="sex")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

```python
# Facet by 'day' and 'sex'
g = sns.FacetGrid(tips, row="sex", col="day")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```
